```python
import faust
```

```python
app = faust.App("HateSpeechDetection", broker='kafka://localhost:9092', value_serializer='raw')
kafka_topic = app.topic("HateSpeechStream", value_type='json')

@app.agent(kafka_topic)
async def print_text(data):
    async for tweet in data:
        print(tweet)
        
```

```python
if __name__ == '__main__':
    app.main()
```

```python
from kafka import KafkaConsumer

import json
from json import loads
```

```python
consumer = KafkaConsumer(bootstrap_servers='localhost:9092', auto_offset_reset='latest', group_id='idaid', consumer_timeout_ms=30000, value_deserializer=lambda x: loads(x.decode('utf-8')))
consumer.subscribe('HateSpeechStream')
```

```python
for message in consumer:
    message = message.value;
    message = json.dumps(message)
    message = json.loads(message)
    try:
        message["retweeted_status"]
        message = message["extended_tweet"]
        message = json.dumps(message)
        message = json.loads(message)
        print(message["full_text"])
        print("------------------------------------")
    except KeyError:
        try:
            print(message["text"])
            print("------------------------------------")
        except KeyError:
            continue
```
